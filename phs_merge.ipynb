```python
lst = []
for staff_number, dept, job, last_name, first_name in zip(
    df_card_data.staff_number.values, df_card_data.dept.values, df_card_data.job.values,
    df_card_data.kanji_last_name.values ,df_card_data.kanji_first_name.values, 
):
    name = ""
    if isinstance(last_name, float)==True and isinstance(first_name, float)==False:
        name = first_name
    elif isinstance(last_name, float)==False and isinstance(first_name, float)==True:
        name = last_name
    elif isinstance(last_name, float) and isinstance(first_name, float):
        name = np.nan
    else:
        name = last_name + first_name
    lst.append([staff_number, dept, job, name])
```

```python
df_staff = pd.DataFrame(lst, columns=["staff_number", "dept", "job", "name"])
df_staff = df_staff.dropna(subset=["staff_number", "dept", "job", "name"])
```
